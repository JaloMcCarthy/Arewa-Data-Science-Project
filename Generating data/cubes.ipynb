#### Exercise 15.1: Cubes

+ A number raised to the third power is a cube. Plot the first five cubic numbers, and then plot the first 5,000 cubic numbers.A number raised to the third power is a cube. Plot the first five cubic numbers, and then plot the first 5,000 cubic numbers.

```python
import matplotlib.pyplot as plt

first_five = [x**3 for x in range(1, 6)]

first_five_thousand = [x**3 for x in range(1, 5001)]

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1, 6), first_five, marker='o', linestyle='-', color='blue', label='First 5 Cubic Numbers')
plt.title('First 5 Cubic Numbers')
plt.xlabel('Number')
plt.ylabel('Cube')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(range(1, 5001), first_five_thousand, linestyle='-', color='green', label='First 5000 Cubic Numbers')
plt.title('First 5,000 Cubic Numbers')
plt.xlabel('Number')
plt.ylabel('Cube')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

#### Exercise 15.2: Coloured cubes
+ Apply a colormap to your cubes plot.

```python
import matplotlib.pyplot as plt
import numpy as np

first_five = [x**3 for x in range(1, 6)]

first_five_thousand = [x**3 for x in range(1, 5001)]

colors_five = plt.cm.viridis(np.linspace(0, 1, len(first_five)))
colors_five_thousand = plt.cm.plasma(np.linspace(0, 1, len(first_five_thousand)))

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
for i, value in enumerate(first_five):
    plt.bar(i + 1, value, color=colors_five[i], label=f'{i + 1}^3 = {value}')
plt.title('First 5 Cubic Numbers')
plt.xlabel('Number')
plt.ylabel('Cube')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(range(1, 5001), first_five_thousand, c=colors_five_thousand, s=1, label='Cubic Values')
plt.title('First 5,000 Cubic Numbers')
plt.xlabel('Number')
plt.ylabel('Cube')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
